```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('auto_mpg_dataset.csv')
```

```python
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,8:].values
```

```python
""" Using Linear Reg."""
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state =0)

from sklearn.linear_model import LinearRegression
lin_model = LinearRegression()
lin_model.fit(X_train,y_train)
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, lin_model.predict(X_test), color = 'blue')
lin_y_pred = lin_model.predict(X_test)
from sklearn.metrics import r2_score
x= r2_score(lin_y_pred,y_test)
print(x)
""" End """
```

```python
"""Using Multi Linear Reg. """
multi_dataset = dataset[dataset['horsepower']>0]
multi_X = multi_dataset.iloc[:, :-2].values
multi_y = multi_dataset.iloc[:,8:].values

import statsmodels.regression.linear_model as lm
multi_X = np.append(np.ones((len(multi_X),1)).astype(int),values = multi_X,axis = 1)

X_opt = multi_X[:,[0,1,2,3,4,5,6,7]]
regressor_ols = lm.OLS(endog = multi_y, exog = X_opt).fit()
regressor_ols.summary()

X_opt = multi_X[:,[0,1,2,3,4,6,7]]
regressor_ols = lm.OLS(endog = multi_y, exog = X_opt).fit()
regressor_ols.summary()

X_opt = multi_X[:,[0,2,3,4,6,7]]
regressor_ols = lm.OLS(endog = multi_y, exog = X_opt).fit()
regressor_ols.summary()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_opt, multi_y, test_size = 1/3, random_state =0)

from sklearn.linear_model import LinearRegression
lin_model_2 = LinearRegression()
lin_model_2.fit(X_train,y_train)
multi_y_pred = lin_model_2.predict(X_test)
from sklearn.metrics import r2_score
x= r2_score(multi_y_pred,y_test)
print(x)
""" End """
```

```python
"""Using poli Linear Reg. """

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
poly_reg.fit(X, y)
lin_model_3 = LinearRegression()
lin_model_3.fit(poly_reg.fit_transform(X), y)

plt.scatter(X, y, color = 'red')
plt.plot(X, lin_model_3.predict(poly_reg.fit_transform(X)), color = 'blue')
poli_y_pred = lin_model_3.predict(poly_reg.fit_transform(X))
from sklearn.metrics import r2_score
x= r2_score(poli_y_pred,y)
print(x)
""" End """
```

```python
"""Using Decision Tree Reg. """

from sklearn.tree import DecisionTreeRegressor
lin_model_4 = DecisionTreeRegressor(random_state = 0)
lin_model_4.fit(X, y)

X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_model_4.predict(X_grid), color = 'blue')
y_pred = lin_model_4.predict(X_grid)
from sklearn.metrics import r2_score
x= r2_score(poli_y_pred,y)
print(x)
""" End """
```

```python
"""Using Random Forest Tree Reg. """
from sklearn.ensemble import RandomForestRegressor
lin_model_4 = RandomForestRegressor(n_estimators = 1000, random_state = 0)
lin_model_4.fit(X, y)
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_model_4.predict(X_grid), color = 'blue')
y_pred = lin_model_4.predict(X_grid)
from sklearn.metrics import r2_score
x= r2_score(poli_y_pred,y)
print(x)
""" End """
```
